```python
import numpy as np
import pandas as pd
```

```python
file_path = r"\\172.16.0.126\raaa shared\1. RAAA - SOLAR\Manjot\july'26\GJ-S CMv profile july'26.xlsx"
```

```python
df = pd.read_excel(file_path)
df[df.select_dtypes(include=['number']).columns] = df.select_dtypes(include=['number']).clip(lower=0)
df.fillna(0, inplace=True)
days = int(len(df)/96)
```

```python
df.head()
```

```python
x = np.array(df['GHI']).reshape(days, 96)
y = np.array(df['Power']/1000).reshape(days, 96)

x = np.percentile(x, 95, axis=0)
y = np.percentile(y, 95, axis=0)
print(x)
```

```python
data1 = {
    'GHI': x,
    'Power': y
}

df2 = pd.DataFrame(data1)
df2.tail()
```

```python
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df2.to_excel(writer, sheet_name='AUG-Percentile')
```
