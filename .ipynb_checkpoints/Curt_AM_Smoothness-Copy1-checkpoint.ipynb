```python
import pandas as pd
import numpy as np
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\Manjot Singh\Desktop\Fatehgarh XL Xergi_T21_14_05_2026_20_05_2026.xlsx"
```

```python
df = pd.read_excel(file_path)
df.fillna(0, inplace=True)
days = int(len(df)/96)
df.head()
```

```python
a = np.array(df["Power (MW)"])

a = a.reshape(days, 96)

ap = np.percentile(a, 95, axis=0)

s = savgol_filter(ap, window_length=7, polyorder=3)
sh = np.roll(s, -2)
sym = (s + sh[::-1]) / 2

df2 = pd.DataFrame({"Power" : ap})
df2["Smooth-Pro"] = sym
df2
```

```python
p = np.arange(0, 96)

plt.figure(figsize=(12,6))
plt.plot(p, sym, label='Sm Avg', color='blue', linewidth=2)
plt.plot(p, ap, label='Avg', color='red', linewidth=2)
plt.xlim(0, 96)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
pars = savgol_filter(s, window_length=7, polyorder=2)
def make_para_curve(
    y,
    peak_cap=300,
    rise_points=8,
    slope_threshold=0.8
):

    y = np.array(y, dtype=float)

    n = len(y)

    para = y.copy()

    # smooth signal
    smooth_y = savgol_filter(y, 7, 2)

    # ===================================
    # FIND SLOPE
    # ===================================

    grad = np.gradient(smooth_y)

    peak_idx = np.argmax(smooth_y)

    # ===================================
    # FIT RISING SECTION
    # ===================================

    start_fit = max(0, peak_idx - rise_points)

    x_rise = np.arange(start_fit, peak_idx)
    y_rise = smooth_y[start_fit:peak_idx]

    coeff = np.polyfit(x_rise, y_rise, 1)

    slope = coeff[0]
    intercept = coeff[1]

    # ===================================
    # EXTRAPOLATE UPWARD
    # ===================================

    rise_curve = []

    idx = peak_idx

    while True:

        val = slope * idx + intercept

        if val >= peak_cap:
            val = peak_cap
            rise_curve.append(val)
            break

        rise_curve.append(val)

        idx += 1

    rise_curve = np.array(rise_curve)

    top_start = peak_idx
    top_end = top_start + len(rise_curve)

    para[top_start:top_end] = rise_curve

    # ===================================
    # AUTO FLAT WIDTH DETECTION
    # ===================================

    flat_width = 0

    for i in range(top_end, n-1):

        if abs(grad[i]) < slope_threshold:
            flat_width += 1
        else:
            break

    # minimum safety
    flat_width = max(flat_width, 3)

    flat_end = top_end + flat_width

    para[top_end:flat_end] = peak_cap

    # ===================================
    # SYMMETRIC FALL
    # ===================================

    fall_curve = rise_curve[::-1]

    fall_len = min(len(fall_curve), n - flat_end)

    para[flat_end:flat_end+fall_len] = fall_curve[:fall_len]

    return para
# ===================================
# USE
# ===================================

parv = make_para_curve(
    pars,
    peak_cap=300,
    rise_points=8,
    slope_threshold=0.8
)

plt.figure(figsize=(14,5))

plt.plot(ap, label="Generation")
plt.plot(parv, linewidth=3, label="Para Curve")

plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
pars = savgol_filter(s, window_length=7, polyorder=2)
def make_para_curve(
    y,
    peak_cap=300,
    rise_points=8,
    flat_width=4
):

    y = np.array(y, dtype=float)

    n = len(y)

    para = y.copy()

    # ===================================
    # FIT RISING REGION
    # ===================================

    peak_idx = np.argmax(y)

    start_fit = max(0, peak_idx - rise_points)

    x_rise = np.arange(start_fit, peak_idx)
    y_rise = y[start_fit:peak_idx]

    # linear fit
    coeff = np.polyfit(x_rise, y_rise, 1)

    slope = coeff[0]
    intercept = coeff[1]

    # ===================================
    # EXTRAPOLATE UPWARD TO 300
    # ===================================

    rise_curve = []

    idx = peak_idx

    while True:

        val = slope * idx + intercept

        if val >= peak_cap:
            val = peak_cap
            rise_curve.append(val)
            break

        rise_curve.append(val)

        idx += 1

    rise_curve = np.array(rise_curve)

    top_start = peak_idx
    top_end = top_start + len(rise_curve)

    # insert rise extrapolation
    para[top_start:top_end] = rise_curve

    # ===================================
    # FLAT TOP
    # ===================================

    flat_start = top_end
    flat_end = flat_start + flat_width

    para[flat_start:flat_end] = peak_cap

    # ===================================
    # AUTOMATIC SYMMETRIC FALL
    # ===================================

    fall_curve = rise_curve[::-1]

    fall_len = min(len(fall_curve), n - flat_end)

    para[flat_end:flat_end+fall_len] = fall_curve[:fall_len]

    return para


# ===================================
# USE
# ===================================

parv = make_para_curve(
    pars,
    peak_cap=250,
    rise_points=8,
    flat_width=24
)
def Perfect(m, peak_val):

    Smooth = savgol_filter(m, window_length=11,polyorder=3)

    # limit values
    Smooth = np.clip(Smooth, 0, peak_val)

    return Smooth


Final_Smooth = Perfect(
    parv,
    peak_val=250
)

df2["Smooth-Pro-Curt"] = Final_Smooth
plt.figure(figsize=(14,5))

plt.plot(ap, label="Generation")
plt.plot(Final_Smooth, linewidth=3, label="Final_Smooth")
#plt.plot(parv, linewidth=3, label="Para Curve")
plt.grid(True)
plt.legend()
plt.show()
```
