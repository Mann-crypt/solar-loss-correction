```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
file_path = r"C:\Users\Manjot Singh\Desktop\GHI Profile (14).xlsx"
```

```python
df = pd.read_excel(file_path)
df.head()
```

```python
actual = np.array(df.iloc[:, 1])
forecast = np.array(df.iloc[:, 2])*1.1
np.argmax(forecast > 5)
```

```python
forecast_start = np.argmax(forecast > 10)

def sigmoid(x, center, steepness):
    return 1 / (1 + np.exp(-(x - center) / steepness))

x = np.arange(len(forecast))

least_error = np.inf
best_center = None

steepness = 3

# Search from a little before sunrise to about 2 hours after
search_start = max(0, forecast_start - 4)
search_end = min(48, forecast_start + 22)

least_error = np.inf
best_center = forecast_start

for center in range(search_start, search_end):

    weight = sigmoid(x, center, steepness)
    corrected = forecast * weight

    end = np.argmax(actual > 100)   # or another suitable threshold

    rmse = np.sqrt(np.mean((actual[:end] - corrected[:end])**2))

    if rmse < least_error:
        least_error = rmse
        best_center = center

print("Best center:", best_center)
print("RMSE:", least_error)

weight = sigmoid(x, best_center, steepness)
corrected = forecast * weight
```
