```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
import seaborn as sns
```

```python
file_path = r"\\172.16.0.126\raaa shared\1. RAAA - SOLAR\Manjot\july'26\GJ-S CMv profile july'26.xlsx"
col_drop = "Avg"
```

```python
df = pd.read_excel(file_path)
df.head(5)
```

```python
min_data_requirement = len(df) * 0.3
df = df.dropna(axis=1, thresh=min_data_requirement)
df = df.fillna(0)
#Constant Values Zero
df = df.copy()

for col in df.columns:
    s = df[col]

    if not pd.api.types.is_numeric_dtype(s):
        continue

    group = (s != s.shift()).cumsum()
    group_size = group.map(group.value_counts())

    df[col] = s.mask((group_size > 2) & (s != 0), 0)
    
days = int(len(df)/96)
#df = df.drop(columns=["Avg"])
df.head(5)
#df = df.set_index('Date')
```

```python
df.info()
```

```python
#Reshape and Percentile
result = {}

for col in df.columns:
    data = df[col].values
    
    reshaped = data.reshape(days, 96)
    perc_values = np.percentile(reshaped, 95, axis=0)
    
    result[col] = perc_values

df2 = pd.DataFrame(result)

df2[col] = df2[col].where(df2[col].diff() != 0, 0)

# Smoothness and Profile
df2['Average'] = df2.mean(axis =1)

a = np.array(df2['Average'])

s = savgol_filter(a, window_length=17, polyorder=3)
s = np.clip(s, 0, None)
s = np.where(s < 0.4, 0, s)
df2['smooth_Average'] = s
df2['smooth_Average'] = df2['smooth_Average'].clip(lower=0)
s = df2['smooth_Average']
df2.head(5)
```

```python
p = np.arange(0, 96)

plt.figure(figsize=(12,6))
plt.plot(p, s, label='Sm Avg', color='blue', linewidth=2)
plt.plot(p, a, label='Avg', color='red', linewidth=2)
plt.xlim(0, 96)
plt.legend()
plt.show()
```

```python
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df2.to_excel(writer, sheet_name='CMV_Curve')
```
