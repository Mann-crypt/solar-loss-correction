```python
#Constant Values Zero
df = df.copy()

for col in df.columns:
    s = df[col]

    if not pd.api.types.is_numeric_dtype(s):
        continue

    group = (s != s.shift()).cumsum()
    group_size = group.map(group.value_counts())

    df[col] = s.mask((group_size > 2) & (s != 0), 0)

raw_x = np.array(df['GHI'])
df.head(5)
```

```python
BLOCKS = 96

for start in range(0, len(df), BLOCKS): 
    
    end = start + BLOCKS 
    
    ghi = df.iloc[start:end]["GHI"].values 
    
    power = df.iloc[start:end]["Power"].values.astype(float) 
    
    ratio = np.divide( power, ghi, out=np.zeros_like(power), where=ghi > 50 ) 
    
    median_ratio = np.median(ratio[ghi > 100]) 
    
    bad = ( (ghi > 250) & (ratio < 0.92 * median_ratio) ) 
    
    power[bad] = 0 
    
    df.iloc[start:end, df.columns.get_loc("Power")] = power
```

```python
x = np.array(df['GHI']).reshape(-1, 1)
y = np.array(df['Power']/1000).reshape(-1, 1)

X = np.column_stack((x,y))

# Standardize first
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)

db = DBSCAN(eps=0.3, min_samples=10)
labels = db.fit_predict(X_scaled)

mask = labels != -1

x = x[mask]
y = y[mask]
len(y)
```

```python
def gompertz(x, L, k, c):
    return L * np.exp(-np.exp(-k * (x - c)))

def richards(x, A, K, B, M, nu):
    return A + (K - A) / ((1 + np.exp(-B * (x - M))) ** (1 / nu))
```

```python
import os
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error

# ---------- INPUT ----------
x = np.array(x).flatten()
y = np.array(y).flatten()
weights = 0.1 + 0.9 * (x.flatten() / x.max())**2

# ---------- CLEAN ----------
data = pd.DataFrame({'x': x, 'y': y}).fillna(0)
x = data['x'].values
y = data['y'].values

results = {}

# ---------- SAFE MAPE FUNCTION ----------
def safe_mape(y_true, y_pred):
    mask = y_true != 0
    return mean_absolute_percentage_error(y_true[mask], y_pred[mask])

# ---------- 1. LINEAR ----------
#model = LinearRegression()
#model.fit(x.reshape(-1,1), y)
#y_pred_linear = model.predict(x.reshape(-1,1))

#results["Linear"] = {
#    "r2": r2_score(y, y_pred_linear),
#    "type" : "linear",
#    "mae": mean_absolute_error(y, y_pred_linear),
#    "mape": safe_mape(y, y_pred_linear),
#    "x": x,
#    "y_pred": y_pred_linear
#}

# ---------- 2. POLY 2 ----------
#poly2 = np.poly1d(np.polyfit(x, y, 2, w=weights))
#y_pred_poly2 = poly2(x)

#results["Poly 2"] = {
#    "model" : poly2,
#    "r2": r2_score(y, y_pred_poly2),
#    "type" : "poly",
#    "mae": mean_absolute_error(y, y_pred_poly2),
#    "mape": safe_mape(y, y_pred_poly2),
#    "x": x,
#    "y_pred": y_pred_poly2
#}
# ---------- 4. GOMPERTZ ----------
try:
    initial_guess = [
        np.max(y),        # Maximum power
        0.01,             # Growth rate
        np.median(x)      # Inflection point
    ]

    params, _ = curve_fit(
        gompertz,
        x,
        y,
        p0=initial_guess,
        maxfev=10000
    )

    y_pred_gomp = gompertz(x, *params)

    results["Gompertz"] = {
        "model": params,
        "params": params,
        "r2": r2_score(y, y_pred_gomp),
        "type": "gompertz",
        "mae": mean_absolute_error(y, y_pred_gomp),
        "mape": safe_mape(y, y_pred_gomp),
        "x": x,
        "y_pred": y_pred_gomp
    }

except RuntimeError:
    print("Gompertz fitting failed.")

#---------- 3. POLY 3 ----------
poly3 = np.poly1d(np.polyfit(x, y, 3, w=weights))
y_pred_poly3 = poly3(x)

results["Poly 3"] = {
    "model" : poly3,
    "r2": r2_score(y, y_pred_poly3),
    "type" : "poly",
    "mae": mean_absolute_error(y, y_pred_poly3),
    "mape": safe_mape(y, y_pred_poly3),
    "x": x,
    "y_pred": y_pred_poly3
}

# ==========================================================
# HYBRID (Poly3 + Gompertz)
# ==========================================================

poly_pred = poly3(x)
gomp_pred = gompertz(x, *params)

# Find where both models are naturally closest

mask = x > 0.5 * x.max()

difference = np.abs(poly_pred[mask] - gomp_pred[mask])

transition = x[mask][np.argmin(difference)]

blend_width = 150

blend_start = transition - blend_width/2
blend_end   = transition + blend_width/2

alpha = np.clip(
    (x - blend_start) /
    (blend_end - blend_start),
    0,
    1
)

hybrid_pred = (
    (1-alpha) * poly_pred +
    alpha * gomp_pred
)

results["Hybrid"] = {
    "model": (poly3, params, transition, blend_width),
    "params": params,
    "transition": transition,
    "blend_width": blend_width,
    "type": "hybrid",
    "r2": r2_score(y, hybrid_pred),
    "mae": mean_absolute_error(y, hybrid_pred),
    "mape": safe_mape(y, hybrid_pred),
    "x": x,
    "y_pred": hybrid_pred
}


    
# ---------- PRINT METRICS ----------
print("\nMODEL COMPARISON:\n")
for name, metrics in results.items():
    print(f"{name}: R²={metrics['r2']:.4f}, "
          f"MAE={metrics['mae']:.4f}, "
          f"MAPE={metrics['mape']*100:.2f}%")

# ---------- NORMALIZATION VALUES ----------
MAE_values = [results[name]['mae'] for name in results]
MAPE_values = [results[name]['mape'] for name in results]

# ---------- SCORING FUNCTION ----------
def score_model(m):
    return (
        m['r2']
        - (m['mae'] / max(MAE_values))
    )

# ---------- BEST MODEL ----------
best_model_name, best_metrics = max(
    results.items(),
    key=lambda item: score_model(item[1])
)

best_pred = best_metrics["y_pred"]
best_x = best_metrics["x"]

best_model = best_metrics["model"]
best_type = best_metrics["type"]

print(f"\nBest Model: {best_model_name}")

# ---------- ADD BEST PREDICTION ----------
data["Best_Prediction"] = best_pred

# ---------- BAR PLOT FUNCTION ----------
def plot_bar(values, title, ylabel, is_percent=False):
    plt.figure(figsize=(6,4))
    bars = plt.bar(model_names, values)

    for i, name in enumerate(model_names):
        if name == best_model_name:
            bars[i].set_linewidth(3)

    for i, v in enumerate(values):
        label = f"{v*100:.2f}%" if is_percent else f"{v:.3f}"
        plt.text(i, v + 0.01, label, ha='center')

    plt.style.use('seaborn-v0_8')
    plt.xticks(rotation=45)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(axis='y')
    plt.show()

# ---------- PLOT METRICS ----------
model_names = list(results.keys())

r2_values = [results[name]['r2'] for name in model_names]
mae_values = [results[name]['mae'] for name in model_names]
mape_values = [results[name]['mape'] for name in model_names]

plot_bar(r2_values, "R² Comparison", "R²")
plot_bar(mae_values, "MAE Comparison", "MAE")
plot_bar(mape_values, "MAPE Comparison", "MAPE (%)", is_percent=True)

# ---------- PLOT ALL MODELS ----------
for name, metrics in results.items():

    x_vals = metrics["x"]
    y_vals = metrics["y_pred"]
    score = metrics["r2"]

    plt.figure(figsize=(10,6))
    plt.scatter(x, y, alpha=0.5, label="Original Data")

    sorted_idx = np.argsort(x_vals)

    if name == best_model_name:
        plt.plot(x_vals[sorted_idx], y_vals[sorted_idx],
                 linewidth=2, linestyle='--', color="yellow",
                 label=f"{name} (BEST, R²={score:.3f})")
    else:
        plt.plot(x_vals[sorted_idx], y_vals[sorted_idx],
                 linewidth=2, color="yellow",
                 label=f"{name} (R²={score:.3f})")

    plt.legend()
    plt.xlabel("GHI")
    plt.ylabel("Power")
    plt.title(f"{name} Trendline")
    plt.grid(True)
    plt.show()
```

```python
rx = raw_x.reshape(days, 96)
rxf = p_f.reshape(day, 96)

px = np.percentile(rx, 95, axis=0)
pxf = np.percentile(rxf, 95, axis=0)

factors = np.arange(0.75, 1.251, 0.01)
best_factor = None
least_error = np.inf
for f in factors:
    
    scaled_curve = pxf * f
    
    # Mean Squared Error
    error = np.mean((px - scaled_curve) ** 2)
    
    if error < least_error:
        least_error = error
        best_factor = f

print("Best multiplier:", best_factor)
print("Least error:", least_error)

n_f = x_f * best_factor
dxf = n_f.reshape(day, 96)
dpxf = np.percentile(dxf, 95, axis=0)
print("GHI :",(np.max(px)),", F_GHI :",np.max(x_f))

# Early Start fix
forecast = dpxf
actual = px
forecast_start = np.argmax(forecast > 10)
actual_start = np.argmax(actual > 10)

print(forecast_start, actual_start+8)

def sigmoid(x, center, steepness):
    return 1 / (1 + np.exp(-(x - center) / steepness))

x = np.arange(len(forecast))

least_loss = np.inf
best_center = None
best_steepness = None

# Search region
for center in range(max(0, forecast_start-5), min(48, forecast_start+20)):

    for steepness in np.arange(0.5, 4.25, 0.25):

        weight = sigmoid(x, center, steepness)
        corrected = forecast * weight

        # 1. First-half RMSE
        rmse_half = np.sqrt(
            np.mean((actual[:48] - corrected[:48])**2)
        )

        # 2. Sunrise RMSE (from forecast start to actual start + few intervals)
        sunrise_end = min(actual_start + 8, 48)

        rmse_start = np.sqrt(np.mean((actual[forecast_start:sunrise_end] - corrected[forecast_start:sunrise_end])**2))

        # Combined loss
        loss = 0.7 * rmse_half + 0.3 * rmse_start

        if loss < least_loss:
            least_loss = loss
            best_center = center
            best_steepness = steepness

print("Best Center:", best_center, ", Best Steepness:", best_steepness, ", Least Loss:", least_loss)
weight = sigmoid(x, best_center, best_steepness)
corrected = forecast * weight

print(len(forecast))

a = np.arange(0, 96)
plt.figure(figsize=(15,5))
plt.xlabel('GHI')
plt.ylabel('Forecast GHI')
plt.xlim(10, 96)   # zoom X-axis
plt.ylim(0, 1100)   # zoom Y-axis
plt.plot(a, pxf, label="forecasted GHI")
plt.plot(a, dpxf, label="calc forecasted GHI")
plt.plot(a, corrected, label="corr forecasted GHI")
plt.plot(a, px, label="GHI")
plt.legend()
plt.show()
```
