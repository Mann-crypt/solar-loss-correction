```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
file_path = r"C:\Users\Manjot Singh\Desktop\GHI Profile (19).xlsx"
```

```python
df = pd.read_excel(file_path)
df.head()
```

```python
actual = np.array(df.iloc[:, 1])
forecast = np.array(df.iloc[:, 4])
```

```python
import numpy as np

forecast_start = np.argmax(forecast > 10)
actual_start = np.argmax(actual > 10)

print(forecast_start, actual_start+8)

def sigmoid(x, center, steepness):
    return 1 / (1 + np.exp(-(x - center) / steepness))

x = np.arange(len(forecast))

least_loss = np.inf
best_center = None
best_steepness = None

# Search region
for center in range(max(0, forecast_start-5), min(48, forecast_start+20)):

    for steepness in np.arange(0.5, 4.25, 0.25):

        weight = sigmoid(x, center, steepness)
        corrected = forecast * weight

        # 1. First-half RMSE
        rmse_half = np.sqrt(
            np.mean((actual[:48] - corrected[:48])**2)
        )

        # 2. Sunrise RMSE (from forecast start to actual start + few intervals)
        sunrise_end = min(actual_start + 8, 48)

        rmse_start = np.sqrt(np.mean((actual[forecast_start:sunrise_end] - corrected[forecast_start:sunrise_end])**2))

        # Combined loss
        loss = 0.7 * rmse_half + 0.3 * rmse_start

        if loss < least_loss:
            least_loss = loss
            best_center = center
            best_steepness = steepness

print("Best Center:", best_center, ", Best Steepness:", best_steepness, ", Least Loss:", least_loss)
weight = sigmoid(x, best_center, best_steepness)
corrected = forecast * weight
```

```python
p = np.arange(0, 96)
plt.figure(figsize=(14,6))
plt.plot(p, corrected, label='correct-forecast', color='green', linewidth=2)
plt.plot(p, forecast, label='forecast', color='blue', linewidth=2)
plt.plot(p, actual, label='actual', color='red', linewidth=2)
plt.xlim(0, 96)
plt.legend()
plt.grid(True)
plt.show()
```

```python
df2 = pd.DataFrame({
    "GHI" : actual,
    "Forecasted GHI" : forecast,
    "Corrected Forecasted GHI" : corrected,
    "Weight" : weight
})
```

```python
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df2.to_excel(writer, sheet_name='Early_start_fix_ECM')
```
