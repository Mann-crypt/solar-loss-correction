```python
import pandas as pd

file_path = "C:/Users/Manjot Singh/Desktop/Lalganj GHI.xlsx"

df = pd.read_excel(file_path)

df.head()
```

```python
df['For Date'] = pd.to_datetime(df['For Date'], dayfirst=True)
df = df.set_index('For Date')

df = df.resample('1h').mean()

#df_interp = df.interpolate(method='linear')

#df_15min = df_interp.resample('15min').asfreq()

df_15min = df.resample('15min').interpolate(method='linear')

df_15min.head(50)
```

```python
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df_15min.to_excel(writer, sheet_name='CMV_Curve')
```
