```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
file_path = r"C:\Users\Manjot Singh\Desktop\GHI Profile (13).xlsx"
```

```python
df = pd.read_excel(file_path)
df.head()
```

```python
actual = np.array(df.iloc[:, 1])
forecast = np.array(df.iloc[:, 2])
print(forecast.shape)
print(actual.shape)
```

```python
import numpy as np

forecast_start = np.argmax(forecast > 10)

def sigmoid(x, center, steepness):
    return 1 / (1 + np.exp(-(x - center) / steepness))

x = np.arange(len(forecast))

least_error = np.inf
best_center = None
best_steepness = None

# Search only around the sunrise region
for center in range(max(0, forecast_start - 4), min(48, forecast_start + 12)):

    for steepness in np.arange(0.5, 3, 0.25):

        weight = sigmoid(x, center, steepness)

        corrected = forecast * weight

        start = forecast_start
        end = np.argmax(forecast)

        error = np.sqrt(np.mean((actual[start:end+1] - corrected[start:end+1])**2))

        # Morning-only optimization (recommended)
        #error = np.sqrt(np.mean((actual[:48] - corrected[:48])**2))

        # If you want full-day optimization, use:
        # error = np.sqrt(np.mean((actual - corrected)**2))

        if error < least_error:
            least_error = error
            best_center = center
            best_steepness = steepness

print(f"Best Center     : {best_center}")
print(f"Best Steepness  : {best_steepness}")
print(f"Least RMSE      : {least_error:.3f}")

print(actual[:end])
print(corrected[:end])
# Final corrected forecast
weight = sigmoid(x, best_center, best_steepness)
corrected = forecast * weight
```

```python
weight
```

```python
p = np.arange(0, 96)
plt.figure(figsize=(14,6))
plt.plot(p, corrected, label='correct-forecast', color='green', linewidth=2)
plt.plot(p, forecast, label='forecast', color='blue', linewidth=2)
plt.plot(p, actual, label='actual', color='red', linewidth=2)
plt.xlim(0, 96)
plt.legend()
plt.grid(True)
plt.show()
```

```python
df2 = pd.DataFrame({
    "GHI" : actual,
    "Forecasted GHI" : forecast,
    "Corrected Forecasted GHI" : corrected,
    "Weight" : weight
})
```

```python
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df2.to_excel(writer, sheet_name='Early_start_fix')
```
